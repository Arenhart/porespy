# imshow

## Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

## Create image:

```python
im = ps.generators.blobs([100, 100, 100])
```

## Display image:

```python
fig, ax = plt.subplots(figsize=[7, 7]);
ps.visualization.imshow(im);
plt.axis(False)
plt.title('default')
```

### ind

```python
fig, ax = plt.subplots(figsize=[7, 7]);
ps.visualization.imshow(im, ind =10 );
plt.axis(False)
plt.title('ind = 10')
```

### axis

```python
fig, ax = plt.subplots(figsize=[7, 7]);
ps.visualization.imshow(im, axis = 1);
plt.axis(False)
plt.title('axis = 1')
```
