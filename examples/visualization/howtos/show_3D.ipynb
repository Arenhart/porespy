# show_3D

## Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

## Create image:

```python
im = ps.generators.blobs([100, 100, 100])
```

## Display image:

```python
x = ps.visualization.show_3D(im)

fig, ax = plt.subplots(figsize=[7, 7]);
plt.axis(False);
plt.imshow(x);
```
