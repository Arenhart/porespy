# set_mpl_style

## Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
```

## Create image:

```python
im = ps.generators.blobs([100, 100])
fig, ax = plt.subplots(figsize=[7, 7]);
plt.imshow(im);
```

## Set plot style:

```python
ps.visualization.set_mpl_style()
fig, ax = plt.subplots(figsize=[7, 7]);
plt.imshow(im);
```
