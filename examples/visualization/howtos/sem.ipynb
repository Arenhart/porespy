# sem

## Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

## Create image:

```python
im = ps.generators.blobs([100, 100, 100])
```

## Display image:

```python
x = ps.visualization.sem(im = im)

fig, ax = plt.subplots(figsize=[7, 7]);
plt.axis(False);
plt.imshow(x);
plt.title('default')
```

### axis

```python
x = ps.visualization.sem(im = im, axis = 1)

fig, ax = plt.subplots(figsize=[7, 7]);
plt.axis(False);
plt.imshow(x);
plt.title('axis = 1')
```
