## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
r = 100
ndim = 2
smooth = True
```

## Apply tool:

```python
im = ps.tools.ps_round(r = r, ndim = ndim, smooth = smooth)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)
```

```python
im = ps.tools.ps_round(r = r, ndim = 3, smooth = smooth)
im = im<0.5
im3d = ps.visualization.show_3D(im)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im3d)
```
