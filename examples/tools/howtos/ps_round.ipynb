# ps_round

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables

```python
r = 100
ndim = 2
smooth = True
```

## Apply tool

```python
im = ps.tools.ps_round(r = r, ndim = ndim, smooth = smooth)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im);
```

```python
im = ps.tools.ps_round(r = r, ndim = 3, smooth = smooth)
fig, ax = plt.subplots(figsize=[7, 7]);
ps.visualization.imshow(im);
plt.axis(False);
```
