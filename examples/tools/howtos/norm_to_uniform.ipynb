# norm_to_uniform

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables

```python
im = ps.generators.blobs(shape=[500, 500], porosity=None, blobiness=[1, 1])
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im);

#default values:
scale = None
```

## Apply tool

```python
im = ps.tools.norm_to_uniform(im = im, scale = scale)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im);
```
