# get_planes

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables

```python
im = ps.generators.blobs([100, 100, 100])

fig, ax = plt.subplots(figsize=[7, 7]);
ps.visualization.imshow(im);
plt.axis(False)

#default values:
squeeze = True
```

## Apply tool

```python
im1,im2,im3 = ps.tools.get_planes(im=im, squeeze = squeeze)
fig, ax = plt.subplots(1, 3, figsize=[15, 15]);
ax[0].axis(False)
ax[0].imshow(im1)
ax[1].axis(False)
ax[1].imshow(im2)
ax[2].axis(False)
ax[2].imshow(im3);
```
