# make_contiguous

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables:

```python
im = np.random.randint(-10,10, [10,10], int)
#im = ps.generators.blobs([200, 200])
print(im)
#im = im*(im >0.5)

plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)

#default values:
keep_zeros = True
```

## Apply tool:

```python
im1 = ps.tools.make_contiguous(im = im, keep_zeros=keep_zeros)
im2 = ps.tools.make_contiguous(im = im, keep_zeros=False)
print ('keep_zeros = True: ', im1)
print ('keep_zeros = False:', im2)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].axis(False)
ax[0].imshow(im1)
ax[0].set_title('keep_zeros = True')
ax[1].axis(False)
ax[1].imshow(im2)
ax[1].set_title('keep_zeros = False')
```
