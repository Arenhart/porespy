# ps_disk

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables:

```python
r = 100
smooth = True
```

## Apply tool:

```python
im = ps.tools.ps_disk(r = r, smooth = smooth)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)
```

```python
im = ps.tools.ps_disk(r = r, smooth = False)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)
```
