## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
xmin=1
ymin=1
zmin=1
xmax=30
ymax=40
zmax=50

bbox3d = [xmin,ymin,zmin,xmax,ymax,zmax]
bbox2d = [xmin,ymin,xmax,ymax]

im = ps.generators.blobs([200, 200])
im3d = ps.generators.blobs([200, 200,200])

im3dgraph = ps.visualization.show_3D(im3d)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].axis(False)
ax[0].imshow(im)
ax[0].set_title('2D image')
ax[1].axis(False)
ax[1].imshow(im3dgraph)
ax[1].set_title('3D image')
```

## Apply tool:

```python
box2d = ps.tools.bbox_to_slices(bbox = bbox2d)
box3d = ps.tools.bbox_to_slices(bbox = bbox3d)

im3dbox = ps.visualization.show_3D(im3d[box3d])

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].axis(False)
ax[0].imshow(im[box2d])
ax[0].set_title('2D')
ax[1].axis(False)
ax[1].imshow(im3dbox)
ax[1].set_title('3D')
```
