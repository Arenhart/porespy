# radial_density

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

Computes radial density function by analyzing the histogram of voxel values in the distance transform. 

Create input image and variables:

```python
skel = ps.generators.blobs(shape=[300, 300, 300])
bins = 10
voxel_size = 1
branch_points = None
iterations = 10
im = ps.filters.prune_branches(skel, branch_points, iterations)
```

Apply metric funtion:

```python
x = ps.metrics.radial_density(im, bins, voxel_size)
```

```python
plt.figure(figsize=[10, 10])
plt.title('Probability Density function')
plt.plot(x.pdf)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
plt.title('Cumulative Density Function')
plt.plot(x.cdf)
plt.show()
```
