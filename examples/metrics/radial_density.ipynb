# radial_density
An example for the use of porespy.metrics.radial_density

Computes radial density function by analyzing the histogram of voxel values in the distance transform. 

### Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

### Create image and variables:

```python
skel = ps.generators.blobs(shape=[300, 300])
bins = 10
voxel_size = 1
branch_points = None
iterations = 10
im = ps.filters.prune_branches(skel, branch_points, iterations)

plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)
```

### Apply metric function:

Output is a tuple with elements R, pdf, cdf, bin_centers, bin_edges, and bin widths

```python
x = ps.metrics.radial_density(im, bins, voxel_size)
```

#

Plot results:

```python
plt.figure(figsize=[10, 10])
plt.title('Probability Density function')
plt.plot(x.pdf)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
plt.title('Cumulative Density Function')
plt.plot(x.cdf)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
plt.bar(x.pdf,x.bin_centers, x.bin_widths)
```

```python
plt.figure(figsize=[10, 10])
plt.bar(x.cdf,x.bin_centers, x.bin_widths)
```
