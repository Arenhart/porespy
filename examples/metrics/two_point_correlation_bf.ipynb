# two_point_correlation_bf

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

Calculates the two-point correlation function using brute-force.

Create image and variables:

```python
skel = ps.generators.blobs(shape=[100, 100, 100])
spacing = 5
branch_points = None
iterations = 10
im = ps.filters.prune_branches(skel, branch_points, iterations)
```

Apply metric function:

```python
x = ps.metrics.two_point_correlation_bf(im, spacing)
print(x)
```

```python
plt.figure(figsize=[10, 10])
plt.title('Probability vs. Distance')
plt.plot(x.distance,x.probability)
plt.show()
```
