# phase_fraction

## Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt

ps.visualization.set_mpl_style()
```

## Create image and variables:

```python
im = ps.generators.blobs(shape=[100, 100],frequency=1, octaves=3, gain = 0.5)
im = ps.filters.snow_partitioning(
    im, dt=None, r_max=4, sigma=0.4, return_all=False, mask=True, randomize=True
)
plt.imshow(im)

#default values:
normed = True
```

## Apply metric function:

Output is an array of the amount of voxels with the specified label.

```python
x = ps.metrics.phase_fraction(im, normed)
xx = ps.metrics.phase_fraction(im, False)

print('normed = True', x)
print('normed = False',xx)
```

#

Plot results:

```python
fig, ax = plt.subplots(1, 2, figsize=[20, 15]);
plt.figure(figsize=[10, 10])
ax[0].set_title("normed = True")
ax[1].set_title("normed = False")
ax[0].plot(x)
ax[1].plot(xx)
```
