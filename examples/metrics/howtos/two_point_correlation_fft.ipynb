### Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt

ps.visualization.set_mpl_style()
```

### Create image:

```python
skel = ps.generators.blobs(shape=[100, 100])
im = ps.filters.prune_branches(skel)

plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)
```

### Apply metric function:

Output is a tuple with elements probability and distance

```python
x = ps.metrics.two_point_correlation_fft(im=im)
```

#

Plot results:

```python
plt.figure(figsize=[10, 10])
plt.title("Probability vs. Distance")
plt.plot(x.distance, x.probability)
plt.show()
```
