# two_point_correlation_fft

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

Calculates the two-point correlation function using fourier transforms.

Create input image:

```python
skel = ps.generators.blobs(shape=[300, 300, 300])
branch_points = None
iterations = 10
im = ps.filters.prune_branches(skel, branch_points, iterations)
```

Apply metric filter:

```python
x = ps.metrics.two_point_correlation_fft(im)
```

```python
plt.figure(figsize=[10, 10])
plt.title('Probability vs. Distance')
plt.plot(x.distance,x.probability)
plt.show()
```
