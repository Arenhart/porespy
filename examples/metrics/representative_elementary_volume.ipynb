# representative_elementary_volume

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

Calculates the porosity of the image as a function subdomain size.  This function extracts a specified number of subdomains of random size, then finds their porosity.

Create image:

```python
skel = ps.generators.blobs(shape=[300, 300, 300])
npoints = 1000
branch_points = None
iterations = 10
im = ps.filters.prune_branches(skel, branch_points, iterations)
```

Apply metric function:

```python
profile = ps.metrics.representative_elementary_volume(im, npoints)
```

```python
plt.figure(figsize=[10, 10])
plt.title('Porosity')
plt.plot(profile.porosity)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
plt.title('Volume')
plt.plot(profile.volume)
plt.show()
```
