# region_surface_areas

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

Extracts the surface area of each region in a labeled image.

Create input image and variables:

```python
skel = ps.generators.blobs(shape=[300, 300])
branch_points = None
iterations = 10
im = ps.filters.prune_branches(skel, branch_points, iterations)
r = ps.filters.snow_partitioning(im, dt=None, r_max=4, sigma=0.4, return_all=True, mask=True, randomize=True)
regions = r.regions
voxel_size = 1
strel = None
```

Apply metric filter:

```python
x = ps.metrics.region_surface_areas(regions, voxel_size, strel)
```

```python
plt.figure(figsize=[10, 10])
plt.title('Region Surface Areas')
plt.plot(x)
plt.show()
```
