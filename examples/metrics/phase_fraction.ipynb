# phase_fraction

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

Calculates the number (or fraction) of each phase in an image.

Create input image and variables:

```python
im = ps.generators.perlin_noise(shape = [300,300], porosity=None, octaves = 3,
                 frequency = 1, persistence = 0.5)
normed=True
im = ps.filters.snow_partitioning(im, dt=None, r_max=4, sigma=0.4, return_all=False,
                      mask=True, randomize=True)
plt.imshow(im)
```

Apply metric filter:

```python
x = ps.metrics.phase_fraction(im, normed)
```

```python
plt.figure(figsize=[10, 10])
plt.title('Phase Fractions')
plt.plot(x)
plt.show()
```
