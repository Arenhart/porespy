# phase_fraction
An example for the use of porespy.metrics.phase_fraction

Calculates the number (or fraction) of each phase in an image.

### Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

### Create image and variables:

```python
im = ps.generators.perlin_noise(shape = [300,300], porosity=None, octaves = 3, frequency = 1, persistence = 0.5)
#im = ps.generators.blobs(shape=[300, 300])
#im = ps.filters.apply_chords(im, spacing=1, axis=0, trim_edges=True, label=False)
normed=False
im = ps.filters.snow_partitioning(im, dt=None, r_max=4, sigma=0.4, return_all=False, mask=True, randomize=True)
plt.imshow(im)
```

### Apply metric function:

Output is an array of the amount of voxels with the specified label.

```python
x = ps.metrics.phase_fraction(im, normed)
print(x)
```

#

Plot results:

```python
plt.figure(figsize=[10, 10])
plt.title('Phase Fractions')
plt.plot(x)
plt.show()
```
