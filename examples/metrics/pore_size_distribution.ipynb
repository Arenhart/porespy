# pore_size_distribution
An example for the use of porespy.metrics.pore_size_distribution

Calculate a pore-size distribution based on the image produced by the ``porosimetry`` or ``local_thickness`` functions.

### Import packages

```python
import porespy as ps
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
```

### Create image:

```python
im = ps.generators.blobs(shape=[300, 300])
im = ps.filters.porosimetry(im, sizes=25, inlets=None, access_limited=True, mode='hybrid', fft=True)
bins = 10
log = True
voxel_size = 1

plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)
```

### Apply metric function:

Output is a tuple with the elements R, pdf, cdf, satn, bin_centers, bin_edges, and bin_widths

```python
x = ps.metrics.pore_size_distribution(im, bins=10, log=True, voxel_size=1)
```

#

Plot results:

```python
plt.figure(figsize=[10, 10])
plt.title('Probability Density Function')
plt.plot(x.pdf)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
plt.title('Cumulative Density Function')
plt.plot(x.cdf)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
plt.bar(x.pdf,x.bin_centers, x.bin_widths)
```

```python
plt.figure(figsize=[10, 10])
plt.bar(x.cdf,x.bin_centers, x.bin_widths)
```

```python
plt.figure(figsize=[10, 10])
plt.bar(x.satn,x.bin_centers, x.bin_widths)
```
