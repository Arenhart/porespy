# find_disconnected_voxels
an example for the use of porespy.filters.find_disconnected_voxels

Finds all of the voxels for pores or solids that are not connected to the edges of the image.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=7, volume_fraction=0.5)

plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)

#default values:
conn = None
```

## Apply filter function:
With different parameters

### conn
for 2D options are 4 or 8
for 3D options are 6 or 26

```python
x = ps.filters.find_disconnected_voxels(im, conn=None)
xx = ps.filters.find_disconnected_voxels(im, 4)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('conn = None')
ax[1].set_title('conn = 4')
```
