# distance_transform_lin
an example for the use of porespy.filters.distance_transform_lin

Replaces void voxels with the distance to the nearest solid voxel on the specified axis.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=15, volume_fraction=0.5)
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)

#default values:
axis=0
mode = "both"
```

## Apply filter function:
With different parameters

### axis

```python
x = ps.filters.distance_transform_lin(im, axis, mode)
xx = ps.filters.distance_transform_lin(im, 1, mode)
fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[0].axis(False)
ax[1].imshow(xx);
ax[1].axis(False)
ax[0].set_title('axis = 0')
ax[1].set_title('axis = 1')
```

### mode

```python
xx = ps.filters.distance_transform_lin(im, axis, "forward")
xxx = ps.filters.distance_transform_lin(im, axis, "reverse")
fig, ax = plt.subplots(1, 3, figsize=[20, 20]);
ax[0].imshow(x);
ax[0].axis(False)
ax[1].imshow(xx);
ax[1].axis(False)
ax[2].imshow(xxx);
ax[2].axis(False)
ax[0].set_title('mode = both')
ax[1].set_title('mode = forward')
ax[2].set_title('mode = reverse')
```
