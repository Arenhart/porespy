# distance_transform_lin
an example for the use of porespy.filters.distance_transform_lin

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Replaces void voxels with the distance to the nearest solid voxel on the specified axis.

Generate image and set values for variables:

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=15, volume_fraction=0.5)
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)
axis=0
```

Apply filter:

```python
x1 = ps.filters.distance_transform_lin(im, axis, mode = "both")
x2 = ps.filters.distance_transform_lin(im, axis, mode = "forward")
x3 = ps.filters.distance_transform_lin(im, axis, mode = "reverse")
fig, ax = plt.subplots(1, 3, figsize=[20, 20]);
ax[0].imshow(x1);
ax[0].axis(False)
ax[1].imshow(x2);
ax[1].axis(False)
ax[2].imshow(x3);
ax[2].axis(False)
```
