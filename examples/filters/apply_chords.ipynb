# apply_chords
an example for the use of porespy.filters.apply_chords

Adds chords in the void space separated by 1 voxel plus the input spacing.The direction of the chords is controlled by the input.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([200, 200])
im = ps.generators.RSA(im, radius=25, volume_fraction=0.5)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)

#default values:
spacing=1
axis=0
trim_edges=True
label=False
```

## Apply filter function:
With different parameters

### spacing

```python
x = ps.filters.apply_chords(im, spacing, axis, trim_edges, label)
xx = ps.filters.apply_chords(im, 3, axis, trim_edges, label)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('spacing = 1')
ax[1].set_title('spacing = 3')
```

### axis

```python
xx = ps.filters.apply_chords(im, spacing, 1, trim_edges, label)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('axis = 0')
ax[1].set_title('axis = 1')
```

### trim_edges

```python
xx = ps.filters.apply_chords(im, spacing, axis, False, label)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('trim_edges = True')
ax[1].set_title('trim_edges = False')
```

### label

```python
xx = ps.filters.apply_chords(im, spacing, axis, trim_edges, True)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('label = False')
ax[1].set_title('label = True')
```
