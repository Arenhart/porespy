```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Adds chords in the void space separated by 1 voxel plus the input spacing.The direction of the chords is controlled by the input.

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=25, volume_fraction=0.3)
im = ps.generators.RSA(im, radius=3, volume_fraction=0.7)
plt.imshow(im)
spacing=1
axis=0
trim_edges=True
label=False
```

Apply filters:

```python
x = ps.filters.apply_chords(im, spacing=1, axis=0, trim_edges=True, label=False)
plt.imshow(x)
```
