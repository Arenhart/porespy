# fill_blind_pores
an example for the use of porespy.filters.fill_blind_pores

Fills pores that aren't connected to the edges of an image.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=5, volume_fraction=0.9)
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)

#default values:
conn = None
```

## Apply filter function:
With different parameters

### conn
for 2D options are 4 or 8
for 3D options are 6 or 26

```python
x = ps.filters.fill_blind_pores(im, conn=None)
xx = ps.filters.fill_blind_pores(im, 4)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('conn = None')
ax[1].set_title('conn = 4')
```
