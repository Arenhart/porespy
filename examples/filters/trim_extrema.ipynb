```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Trims local extrema (determined using the mode variable) by and amount specified by h.

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=50, volume_fraction=0.9)
im = ps.filters.distance_transform_lin(im, axis = 0, mode = "reverse")
plt.imshow(im)
h = .99
mode="minima"
```

Apply filter:

```python
x = ps.filters.trim_extrema(im, h, mode)
plt.imshow(x)
```
