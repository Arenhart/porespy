```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Adds padding to an input image, then applys the function selected by "func".

creating example image and variables:

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=20, volume_fraction=0.2)
im = ps.generators.RSA(im, radius=10, volume_fraction=0.4)
im = im == 0
pad_width = 10
pad_val=1
```

using skimage.morphology.skeletonize_3d as the func input:

```python
x =ps.filters.apply_padded(im, pad_width, skimage.morphology.skeletonize_3d, pad_val)
```

```python
plt.imshow(x)
```
