# find_dt_artifacts
an example for the use of porespy.filters.find_dt_artifacts

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Finds points on a distance transform that are closer to wall than solid.

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=15, volume_fraction=0.5)
axis=0
dt = ps.filters.distance_transform_lin(im, axis, mode = "reverse")
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(dt)
```

Apply filters:

```python
x = ps.filters.find_dt_artifacts(dt)
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(x)
```
