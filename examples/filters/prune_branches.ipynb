# prune_branches
an example for the use of porespy.filters.prune_branches

Removes dangling ends or tails of a skeleton.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([300, 300])
skel = ps.generators.RSA(im, radius=25, volume_fraction=0.9)
skel = ps.generators.RSA(skel, radius=10, volume_fraction=0.7)
skel = ps.generators.RSA(skel, radius=5, volume_fraction=0.8)
skel = ps.generators.RSA(skel, radius=3, volume_fraction=0.9)
skel = ps.generators.RSA(skel, radius=2, volume_fraction=0.9)
skel = ps.generators.RSA(skel, radius=1, volume_fraction=0.9)
skel = skel <0.5
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(skel)

#default values:
branch_points = None
iterations = 1
```

## Apply filter function:
With different parameters

### branch_points

```python
x = ps.filters.prune_branches(skel, branch_points, iterations)
bp = ps.generators.RSA(im, radius=75, volume_fraction=0.5)
xx = ps.filters.prune_branches(skel, bp, iterations)

fig, ax = plt.subplots(2, 2, figsize=[15, 15])
ax[0,1].imshow(bp);
ax[0,0].axis(False)
ax[0,1].axis(False)
ax[0,1].set_title('branch_points')
ax[1,0].imshow(x);
ax[1,1].imshow(xx);
ax[1,0].axis(False)
ax[1,1].axis(False)
ax[1,0].set_title('results with branch_points = None')
ax[1,1].set_title('results with branch_points')
```

### iterations

```python
xx = ps.filters.prune_branches(skel, branch_points, 10)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('iterations = 1')
ax[1].set_title('iterations = 10')
```
