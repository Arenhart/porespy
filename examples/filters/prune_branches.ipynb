# prune_branches
an example for the use of porespy.filters.prune_branches

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Removes dangling ends or tails of a skeleton.

Generate input image and variables:

```python
skel = np.zeros([300, 300])
skel = ps.generators.RSA(skel, radius=25, volume_fraction=0.5)
skel = ps.generators.RSA(skel, radius=10, volume_fraction=0.7)
skel = ps.generators.RSA(skel, radius=5, volume_fraction=0.9)
skel = ps.generators.RSA(skel, radius=3, volume_fraction=0.9)
skel = ps.generators.RSA(skel, radius=2, volume_fraction=0.9)
skel = ps.generators.RSA(skel, radius=1, volume_fraction=0.9)
skel = skel <0.5
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(skel)
branch_points = None
iterations = 10
```

Apply filters:

```python
x = ps.filters.prune_branches(skel, branch_points, iterations)

plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(x)
```
