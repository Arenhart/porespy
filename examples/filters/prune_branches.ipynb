# prune_branches
an example for the use of porespy.filters.prune_branches

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Removes dangling ends or tails of a skeleton.

Generate input image and variables:

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=25, volume_fraction=0.3)
skel = ps.generators.RSA(im, radius=3, volume_fraction=0.7)
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(skel)
branch_points = None
iterations = 1
```

Apply filters:

```python
x = ps.filters.prune_branches(skel, branch_points=None, iterations=1)
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(x)
```
