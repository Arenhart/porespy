# apply_chords

## Import packages

```python
import matplotlib.pyplot as plt
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import skimage

ps.visualization.set_mpl_style()
```

## Generate image for testing

```python
im = ps.generators.blobs(shape=[500, 500])
plt.figure(figsize=[4, 4])
plt.axis(False)
plt.imshow(im)

spacing = 1
axis = 0
trim_edges = True
label = False
```

## Apply filter function

### spacing

```python
x = ps.filters.apply_chords(im=im,
                            spacing=spacing,
                            axis=axis,
                            trim_edges=trim_edges,
                            label=label)
xx = ps.filters.apply_chords(im=im,
                             spacing=3,
                             axis=axis,
                             trim_edges=trim_edges,
                             label=label)

fig, ax = plt.subplots(1, 2, figsize=[8, 4])
ax[0].imshow(x)
ax[1].imshow(xx)
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('spacing = 1')
ax[1].set_title('spacing = 3');
```

### axis

```python
xx = ps.filters.apply_chords(im=im,
                             spacing=spacing,
                             axis=1,
                             trim_edges=trim_edges,
                             label=label)

fig, ax = plt.subplots(1, 2, figsize=[8, 4])
ax[0].imshow(x)
ax[1].imshow(xx)
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('axis = 0')
ax[1].set_title('axis = 1');
```

### trim_edges

```python
xx = ps.filters.apply_chords(im=im,
                             spacing=spacing,
                             axis=axis,
                             trim_edges=False,
                             label=label)

fig, ax = plt.subplots(1, 2, figsize=[8, 4])
ax[0].imshow(x)
ax[1].imshow(xx)
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('trim_edges = True')
ax[1].set_title('trim_edges = False');
```

### label

```python
xx = ps.filters.apply_chords(im=im,
                             spacing=spacing,
                             axis=axis,
                             trim_edges=trim_edges,
                             label=True)

fig, ax = plt.subplots(1, 2, figsize=[8, 4])
ax[0].imshow(x)
ax[1].imshow(xx)
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('label = False')
ax[1].set_title('label = True');
```
