# apply_chords

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables:

```python
im = ps.generators.blobs(shape=[500, 500])
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)

#default values:
spacing=1
axis=0
trim_edges=True
label=False
```

## Apply filter function:
With different parameters

### spacing

```python
x = ps.filters.apply_chords(im =im, spacing=spacing, axis=axis, trim_edges=trim_edges, label=label)
xx = ps.filters.apply_chords(im=im, spacing=3, axis=axis, trim_edges=trim_edges, label=label)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('spacing = 1')
ax[1].set_title('spacing = 3')
```

### axis

```python
xx = ps.filters.apply_chords(im=im, spacing=spacing, axis=1, trim_edges=trim_edges, label=label)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('axis = 0')
ax[1].set_title('axis = 1')
```

### trim_edges

```python
xx = ps.filters.apply_chords(im=im, spacing=spacing, axis=axis, trim_edges=False, label= label)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('trim_edges = True')
ax[1].set_title('trim_edges = False')
```

### label

```python
xx = ps.filters.apply_chords(im=im, spacing=spacing, axis=axis, trim_edges=trim_edges, label=True)

fig, ax = plt.subplots(1, 2, figsize=[20, 20]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('label = False')
ax[1].set_title('label = True')
```
