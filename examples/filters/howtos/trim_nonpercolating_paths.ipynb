# trim_nonpercolating_paths

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
ps.visualization.set_mpl_style()
```

## Create image and variables

```python
im = ps.generators.blobs(shape=[500, 500])
plt.figure(figsize=[7,7])
plt.axis(False)
plt.imshow(im)

#default values:
inlet_axis=0
outlet_axis=0
inlets=None
outlets=None
```

## Apply filter function

### inlet_axis

```python
x = ps.filters.trim_nonpercolating_paths(im=im, inlet_axis=inlet_axis, outlet_axis=outlet_axis, inlets=inlets, outlets= outlets)
xx = ps.filters.trim_nonpercolating_paths(im=im, inlet_axis=1, outlet_axis=outlet_axis, inlets=inlets, outlets=outlets)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('inlet_axis = 0')
ax[1].set_title('inlet_axis = 1');
```

### outlet_axis

```python
xx = ps.filters.trim_nonpercolating_paths(im=im, inlet_axis=inlet_axis, outlet_axis=1, inlets=inlets, outlets= outlets)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('outlet_axis = 0')
ax[1].set_title('outlet_axis = 1');
```

### inlet

```python
in2 = ps.generators.blobs([500,500])
xx = ps.filters.trim_nonpercolating_paths(im=im, inlet_axis=inlet_axis, outlet_axis=outlet_axis, inlets=in2, outlets= outlets)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('inlets = None')
ax[1].set_title('inlets = Generated Blobs');
```

### outlets

```python
out2 = ps.generators.blobs([500,500])
xx = ps.filters.trim_nonpercolating_paths(im=im, inlet_axis=inlet_axis, outlet_axis=outlet_axis, inlets=inlets, outlets=out2)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('outlets = None')
ax[1].set_title('outlets = Generated Blobs');
```
