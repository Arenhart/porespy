```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

trim_small_clusters removes clusters from images if they are smaller than the size given as an input.

generate input images:

```python
im = np.zeros([300, 300])
im1 = ps.generators.RSA(im, radius=5, volume_fraction=0.5)
im2 = ps.generators.RSA(im, radius=4, volume_fraction=0.5)
im3 = ps.generators.RSA(im, radius=3, volume_fraction=0.5)
im4 = ps.generators.RSA(im, radius=2, volume_fraction=0.5)
im5 = ps.generators.RSA(im, radius=1, volume_fraction=0.5)


fig, ax = plt.subplots(1, 5, figsize=[20, 20]);
ax[0].imshow(im1);
ax[1].imshow(im2);
ax[2].imshow(im3);
ax[3].imshow(im4);
ax[4].imshow(im5);
```

apply filter:

```python
size=10

x1 = ps.filters.trim_small_clusters(im1, size=10)
x2 = ps.filters.trim_small_clusters(im2, size=10)
x3 = ps.filters.trim_small_clusters(im3, size=10)
x4 = ps.filters.trim_small_clusters(im4, size=10)
x5 = ps.filters.trim_small_clusters(im5, size=10)

fig, ax = plt.subplots(1, 5, figsize=[20, 20]);
ax[0].imshow(x1);
ax[1].imshow(x2);
ax[2].imshow(x3);
ax[3].imshow(x4);
ax[4].imshow(x5);
```
