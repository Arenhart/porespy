# trim_nonpercolating_paths
an example for the use of porespy.filters.trim_nonpercolating_paths

Removes non-percolating paths.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = ps.generators.blobs([300,300])
plt.figure(figsize=[10, 10])
plt.axis(False)
plt.imshow(im)

#default values:
inlet_axis=0
outlet_axis=0
inlets=None
outlets=None
```

## Apply filter function:
With different parameters

### inlet_axis

```python
x = ps.filters.trim_nonpercolating_paths(im, inlet_axis, outlet_axis, inlets, outlets)
xx = ps.filters.trim_nonpercolating_paths(im, 1, outlet_axis, inlets, outlets)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('inlet_axis = 0')
ax[1].set_title('inlet_axis = 1')
```

### outlet_axis

```python
xx = ps.filters.trim_nonpercolating_paths(im, inlet_axis, 1, inlets, outlets)

fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(x);
ax[1].imshow(xx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('outlet_axis = 0')
ax[1].set_title('outlet_axis = 1')
```
