```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

Removes non-percolating paths.

```python
im = np.zeros([300, 300])
im = ps.generators.RSA(im, radius=5, volume_fraction=0.9)
plt.imshow(im)
inlet_axis=0
outlet_axis=0
inlets=None
outlets=None
```

Apply filter:

```python
x = ps.filters.trim_nonpercolating_paths(im, inlet_axis, outlet_axis, inlets, outlets)
plt.imshow(x)
```
