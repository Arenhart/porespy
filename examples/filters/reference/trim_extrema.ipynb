# trim_extrema
``trim_extrema`` is a function which trims local extremes in the greyscale values of an image. It is useful for removing peaks or floods in an image.

## Import Packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Generate Test Image
An image is generated using the ``blobs`` generator with shape of 500 by 500. The ``distance_transform_lin`` filter is applied to this image to create an image with local extremes. This filter works by replacing each voxel with the linear distance to the nearest solid. The final filtered image is used to demonstrate how to trim local extremes and it is visualized below.

```python
im = ps.generators.blobs(shape=[500, 500])
im = ps.filters.distance_transform_lin(im, axis=1, mode='reverse')

plt.figure(figsize=[6, 6])
plt.imshow(im);
plt.axis(False);
```

## Apply Filter
Apply the ``trim_extrema`` filter to the test image. Two different images, x1 and x2, were created by the filter. The ``maxima`` mode is used to create both images specifying the removal of maxima. Different amounts are removed from extrema and the amount is specified by h. The resulting images are visualized removing both h=25 and h=10.

```python
x1 = ps.filters.trim_extrema(im=im, h=25, mode='maxima')
x2 = ps.filters.trim_extrema(im=im, h=10, mode='maxima')

fig, ax = plt.subplots(1, 2, figsize=[12, 12]);
ax[0].imshow(x1);
ax[1].imshow(x2);
ax[0].axis(False);
ax[1].axis(False);
ax[0].set_title('h = 25', fontdict={'fontsize': 18});
ax[1].set_title('h = 10', fontdict={'fontsize': 18});
```

## Modes Available
The available modes are maxima, minima, and extrema. These modes specify which extremes to remove. Minima and maxima modes trim the maximum and minimum extremes respectively. The extrema mode trims both extremes. Three images are created and visualized using each of the three modes.

```python
x3 = ps.filters.trim_extrema(im=im, h=25, mode= 'maxima')
x4 = ps.filters.trim_extrema(im=im, h=25, mode= 'minima')
x5 = ps.filters.trim_extrema(im=im, h=25, mode='extrema')

fig, ax = plt.subplots(1, 3, figsize=[18, 18]);
ax[0].imshow(x3);
ax[1].imshow(x4);
ax[2].imshow(x5);
ax[0].axis(False);
ax[1].axis(False);
ax[2].axis(False);
ax[0].set_title('maxima', fontdict={'fontsize': 18});
ax[1].set_title('minima', fontdict={'fontsize': 18});
ax[2].set_title('extrema', fontdict={'fontsize': 18});
```
