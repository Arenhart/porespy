# `hold_peaks`

Replaces each voxel with the last peak seen along the given axis

```python
import numpy as np
import porespy as ps
import matplotlib.pyplot as plt
from edt import edt
```

## `im`
The input image will most likely be the distant transform

```python
im = ps.generators.blobs(shape=[500, 500])
dt = edt(im)

pk = ps.filters.hold_peaks(im=dt, axis=0)

fig, ax = plt.subplots(1, 2, figsize=[16, 8])
ax[0].imshow(dt/im)
ax[0].axis(False)
ax[0].set_title('distance transform')

ax[1].imshow(pk/im)
ax[1].axis(False)
ax[1].set_title('peaks');
```

## `axis`
Controls the axis and direction of the search, with positive values meaning that scan proceeds from the beginning of the axis and proceed to the end, and negative values meaning vice-versa.

```python
pk1 = ps.filters.hold_peaks(im=dt, axis=1)
pk2 = ps.filters.hold_peaks(im=dt, axis=-1)

fig, ax = plt.subplots(1, 2, figsize=[12, 6])
ax[0].imshow(pk1/im)
ax[0].axis(False)
ax[0].set_title('axis = -1')

ax[1].imshow(pk2/im)
ax[1].axis(False)
ax[1].set_title('axis = 0');
```
