# lattice_spheres

lattice_spheres creates a lattice (of a size determined by the shape input) of spheres with the radius and offset input to the function. The lattice input determines the structure of the lattice; 'sc' is for a simple cubic lattice, 'fcc' is for a face centered cubic latice, and 'bcc' is for a body centered cubic latice.

## Import packages

```python
import numpy as np
import porespy as ps
import scipy
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
np.random.seed(10)
```

## Create variables:

```python
shape = [100,100,100]
radius = 10

#Default Values:
offset = 0
lattice = "sc"
```

## Apply generator function:
With different parameters

### radius

```python
x = ps.generators.lattice_spheres(shape, radius, offset, lattice)
xx = ps.generators.lattice_spheres(shape, 15, offset, lattice)

graphx = ps.visualization.show_3D(x)
graphxx = ps.visualization.show_3D(xx)
fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(graphx);
ax[1].imshow(graphxx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('radius = 10')
ax[1].set_title('radius = 15')
```

### offset

```python
xx = ps.generators.lattice_spheres(shape, radius, 5, lattice)

graphxx = ps.visualization.show_3D(xx)
fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(graphx);
ax[1].imshow(graphxx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('offset = 0')
ax[1].set_title('offset = 5')
```

### lattice
lattice options are sc, fcc, and bcc

```python
xx = ps.generators.lattice_spheres(shape, radius, offset, "fcc")

graphxx = ps.visualization.show_3D(xx)
fig, ax = plt.subplots(1, 2, figsize=[15, 15]);
ax[0].imshow(graphx);
ax[1].imshow(graphxx);
ax[0].axis(False)
ax[1].axis(False)
ax[0].set_title('lattice = sc')
ax[1].set_title('lattice = fcc')
```
