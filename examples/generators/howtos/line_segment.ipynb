# line_segment

## Import packages

```python
import numpy as np
import porespy as ps
import scipy
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
np.random.seed(10)
```

## Create input points:

### 2D example

```python
X02d = [1,200]
X12d = [100,1]

_2dx = [X02d[0],X12d[0]]
_2dy = [X02d[1],X12d[1]]

plt.figure(figsize=[7, 7])
plt.plot(_2dx, _2dy, 'o', color='black');
```

### 3D example

```python
X03d = [50,10,100]
X13d = [200,100,300]

_3dx = [X03d[0],X13d[0]]
_3dy = [X03d[1],X13d[1]]
_3dz = [X03d[2],X13d[2]]

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(_3dx, _3dy, _3dz, 'o', color='black');
```

## Apply generator function

### 2D example

```python
[_2dx,_2dy] = ps.generators.line_segment(X0=X02d, X1=X12d)

plt.figure(figsize=[7, 7])
plt.plot(_2dx, _2dy, 'o', color='black');
```

### 3D example

```python
_3dx,_3dy,_3dz = ps.generators.line_segment(X0=X03d, X1=X13d)

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(_3dx, _3dy,_3dz,'o', color='black');
```
