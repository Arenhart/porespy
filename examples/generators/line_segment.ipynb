# line_segment

```python
import numpy as np
import porespy as ps
import scipy
import matplotlib.pyplot as plt
ps.visualization.set_mpl_style()
np.random.seed(10)
```

line_segment lists points of a line segment between the two points, X0 and X1.

## 2D example

Input points:

```python
X0 = [1,200]
X1 = [100,1]

_2dx = [X0[0],X1[0]]
_2dy = [X0[1],X1[1]]

plt.figure(figsize=[7, 7])
plt.plot(_2dx, _2dy, 'o', color='black');
```

Generate line segment:

```python
[_2dx,_2dy] = ps.generators.line_segment(X0, X1)

plt.figure(figsize=[7, 7])
plt.plot(_2dx, _2dy, 'o', color='black');
```

## 3D example

Input points:

```python
X0 = [50,10,100]
X1 = [200,100,300]

_3dx = [X0[0],X1[0]]
_3dy = [X0[1],X1[1]]
_3dz = [X0[2],X1[2]]

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(_3dx, _3dy, _3dz, 'o', color='black')
```

Generate line segment:

```python
_3dx,_3dy,_3dz = ps.generators.line_segment(X0, X1)

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(_3dx, _3dy,_3dz,'o', color='black')
```
