## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([200, 200,200])
im = ps.generators.RSA(im, radius=20, volume_fraction=0.9)
im = im < 0.5
im3d = ps.visualization.show_3D(im)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im3d)

#default values:
xyz0=[25,25,25]
xyz1=[175,25,175]
r=15
```

## Apply tool:

```python
im = ps.tools.insert_cylinder(im=im, xyz0=xyz0, xyz1=xyz1, r=r)

im3d = ps.visualization.show_3D(im)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im3d)
```
