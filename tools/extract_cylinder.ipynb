## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = ps.generators.blobs([200, 200, 200])
plt.figure(figsize=[7, 7])
plt.axis(False)
im3d = ps.visualization.show_3D(im)
plt.imshow(im3d)

#default values:
r = 50
axis = 0
```

## Apply tool:

```python
im = ps.tools.extract_cylinder(im = im, r = r, axis = axis)
im = im<0.5
plt.figure(figsize=[7, 7])
plt.axis(False)
im3d = ps.visualization.show_3D(im)
plt.imshow(im3d)
```
