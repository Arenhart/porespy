## Import packages

```python
import numpy as np
import porespy as ps
import scipy.ndimage as spim
import matplotlib.pyplot as plt
import skimage
```

## Create image and variables:

```python
im = np.zeros([200, 200])
im = ps.generators.RSA(im, radius=25, volume_fraction=0.5)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)

im = im.astype(int)

#default values:
labels= [1,3,5]
trim = True
```

```python
regions = spim.label(im)[0]
print(regions)

plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(regions)
```

## Apply tool:

```python
im = ps.tools.extract_regions(regions = regions, labels = labels, trim = trim)
plt.figure(figsize=[7, 7])
plt.axis(False)
plt.imshow(im)
```
